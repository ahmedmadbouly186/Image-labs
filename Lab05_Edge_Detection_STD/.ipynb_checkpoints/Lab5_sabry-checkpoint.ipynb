# Reports
- Roberts:
    noisy image produced as size of filter is small.
    
- Prewill:
    less noisy image produced as size of filter is large and 8 filters are used so.
    but still noisy as orginal image has noise.
- Sobel:
    less noisy image produced as size of filter is large and 2 filters are used so.
    but still noisy as orginal image has noise.
- Canny: Best
    non noise image produced, edges are well connected 
    but still noise can be removed with higher sigma for better edge detection

### Canny low_threshold & high_threshold
- low_threshold :
    Lower bound for hysteresis thresholding (linking edges).
    low_threshold is set to 10% of max by default.
- high_threshold : 
    Upper bound for hysteresis thresholding (linking edges).
    high_threshold is set to 20% of max by default.

```python
# 2. Custom Sobel (V-H-Both) using conv2d

hx=np.array([ 
    [-1,-2,-1],
    [0,0,0],
    [1,2,1] 
])
hy=np.array([ 
    [-1,0,1],
    [-2,0,2],
    [-1,0,1] 
])
sobel_x = convolve2d(img,hx,boundary='symm')
sobel_y = convolve2d(img,hy,boundary='symm')
sobelImg = np.sqrt( (sobel_x)**2  +  (sobel_y)**2 )
print("Before Applying thresholding")
show_images([img,sobelImg,sobel_y,sobelImg],["original","sobel_x","sobel_y","sobel"])
#sobel[sobel <= 85] = 0
# sobel = sobel >= 85
sobelImg = sobelImg >= threshold_otsu(sobelImg)
print("after Applying thresholding")
show_images([img,sobel_x,sobel_y,sobelImg],["original","sobel_x","sobel_y","sobel"])
```

```python
def sobelEdgeDetection(img):
    hx = np.array([ 
        [-1,-2,-1],
        [0,0,0],
        [1,2,1] 
    ])
    hy = np.array([ 
        [-1,0,1],
        [-2,0,2],
        [-1,0,1] 
    ])
    sobel_x = np.zeros((img.shape[0],img.shape[1]))
    sobel_y = np.zeros((img.shape[0],img.shape[1]))
    sobelImg = np.zeros((img.shape[0],img.shape[1]))
    for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
            sobel_x[i,j] = np.sum(img[i-1:i+2,j-1:j+2] * hx )
            sobel_y[i,j] = np.sum(img[i-1:i+2,j-1:j+2] * hy )
            sobelImg[i,j] =np.sqrt( (sobel_x[i,j])**2  +  (sobel_y[i,j])**2 )
    #sobel[sobel <= 25] = 0
    sobelImg = sobelImg > 85
    show_images([img,sobel_x,sobel_y,sobelImg],["original","sobel_x","sobel_y","sobel"])
    return sobelImg,sobel_x,sobel_y
```

```python
sobelImg,sobel_x,sobel_y = sobelEdgeDetection(img)
```

```python
# 3. Custom LoG
def LOG(img,thres= 20):
#     print(img)
    h4 = np.array([ 
        [0,1,0],
        [1,-4,1],
        [0,1,0] 
    ])
    h8 = np.array([ 
        [-1,-1,-1],
        [-1,8,-1],
        [-1,-1,-1] 
    ])
    newImg = gaussian(img, sigma=1.2,mode='nearest', cval=0, preserve_range=False, truncate=4.0)
    laplace_4 = np.zeros((img.shape[0],img.shape[1]))
    laplace_8 = np.zeros((img.shape[0],img.shape[1]))
    for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
#             print(newImg[i-1:i+2,j-1:j+2])
            laplace_4[i,j] = np.sum(newImg[i-1:i+2,j-1:j+2] * h4 )
            laplace_8[i,j] = np.sum(newImg[i-1:i+2,j-1:j+2] * h8 )
#     laplace_4 = abs(convolve2d(newImg,h4,boundary='symm'))
#     laplace_8 = abs(convolve2d(newImg,h8,boundary='symm'))
#     laplace_4[laplace_4 < thres] = 0
#     laplace_8[laplace_8 < thres] = 0
    test = laplace(newImg, ksize=3, mask=None)
    laplace_4 = laplace_4 > thres
    laplace_8 = laplace_8 >thres
    test = test > thres
   
    show_images([img,newImg,laplace_4,laplace_8,test],["original","filtered","laplace 4","laplace 8","test"])
    return laplace_4,laplace_8
laplace_4,laplace_8 = LOG(img,0.015)
```

```python
from skimage.filters import laplace
import numpy as np
newImg = gaussian(img, sigma=1,mode='nearest', cval=0, preserve_range=False, truncate=4.0)
lap = np.sign(laplace(newImg))
lap = np.pad(lap, ((0, 1), (0, 1)))
diff_x = lap[:-1, :-1] - lap[:-1, 1:] < 0
diff_y = lap[:-1, :-1] - lap[1:, :-1] < 0

edges =  np.logical_or(diff_x, diff_y).astype(float)
edges = edges > threshold_otsu(edges)
show_images([edges])
```
