```python
'''
1) Erosion / Dilation:
    - Load "coins.jpg" from the img folder
    - Convert the image to binary
    - Write 2 functions each applying 1 of the erosion/dilation algorithms (with adjustable square window size) manually to the binary image
    - Use Skimage's "binary_erosion" & "binary_dilation" & show the results
'''

def gray_scale(image):
    gray_scale_image=image
    if(len(image.shape)==3):
        if(image.shape[2]==4):
            gray_scale_image=rgb2gray(rgba2rgb(image))
        else :
            gray_scale_image=rgb2gray(image)
    else :
        gray_scale_image=gray_scale_image/255
    return gray_scale_image
def erosion(img,SE=np.ones((3,3))):
    erosion_image=np.zeros((img.shape[0],img.shape[1]))
    for x in range(len(img)):
    
        if(x==img.shape[0]-SE.shape[0]-1):
            break
        for y in range(len(img[0])):
            if(y==img.shape[1]-SE.shape[1]-1):
                break
            window=np.zeros((SE.shape[0],SE.shape[1]))
            window=img[x:x+SE.shape[0],y:y+SE.shape[1]]
            
            erosion_image[x+int(SE.shape[0]/2)][y+int(SE.shape[1]/2)]=((SE*window)==SE).all()
    return erosion_image


def dilation(img,SE=np.ones((3,3))):
    dilation_image=np.zeros((img.shape[0],img.shape[1]))
    for x in range(len(img)):
    
        if(x==img.shape[0]-SE.shape[0]-1):
            break
        for y in range(len(img[0])):
            if(y==img.shape[1]-SE.shape[0]-1):
                break
            window=np.zeros((SE.shape[0],SE.shape[1]))
            window=img[x:x+SE.shape[0],y:y+SE.shape[1]]
            
            dilation_image[x+int(SE.shape[0]/2)+1][y+int(SE.shape[1]/2)+1]=((SE[:,:] * window[:,:])==SE).any()

    return dilation_image



origenal_image=io.imread('./img/coins.jpg')

origenal_image_gray=gray_scale(io.imread('./img/coins.jpg'))
origenal_image_binary=origenal_image_gray.copy()

origenal_image_binary[origenal_image_binary<.5]=0
origenal_image_binary[origenal_image_binary>=.5]=1

show_images([origenal_image,origenal_image_gray,origenal_image_binary],['origenal','gray scaled','binary'])

origenal_image_erosin=binary_erosion(origenal_image_binary,np.ones((3,3)))
origenal_image_dilation=binary_dilation(origenal_image_binary)

origenal_image_erosin_func = erosion(origenal_image_binary,np.ones((3,3)))
origenal_image_dilation_func = dilation(origenal_image_binary,np.ones((3,3)))

show_images([origenal_image_erosin,origenal_image_dilation],['erosin','ditation'])
show_images([origenal_image_erosin_func,origenal_image_dilation_func],['handmade erosin','handmade dilation'])

print((origenal_image_dilation_func==origenal_image_dilation).all)
```

```python
'''
2) Case Study: Credit Card Number Extraction
    - Load "card.jpg" from the img folder
    - Convert the image to binary by brightness thresholding maintaining the card number details for extraction
    - Use your erosion/dilation functions, or Skimage's functions only if you didn't implement them, to apply closing
        Note: The goal of this step is to join the card number into 4 sets of connected pixels
    - Use Skimage's "find_contours" to get the bounding boxes for the credit card's numbers by filtering on aspect ratio (between 2.5 & 3.5)
        Note: When using "find_contours(img, 0.8)" the o/p is a list of contours each having:
                - contour[:,0] being the Y values of the contour perimeter points
                - contour[:,1] being the X values of the contour perimeter points
'''
origenal_image=io.imread('./img/card.jpg')
origenal_image_gray=gray_scale(io.imread('./img/card.jpg'))
origenal_image_binary=origenal_image_gray.copy()
origenal_image_binary[origenal_image_binary<.65]=0
origenal_image_binary[origenal_image_binary>=.65]=1
show_images([origenal_image_gray,origenal_image_binary],['gray scaled','binary'])

closed_image=origenal_image_binary.copy()
closed_image=dilation(closed_image,np.ones((13,13)))
closed_image=erosion(closed_image,np.ones((13,13)))

show_images([closed_image],'closed')


```

```python
contours=find_contours(closed_image,.8)

bounding_boxes=[]
for contour in contours[1:]: 
    x_length=round(max(contour[:,1]))-round(min(contour[:,1]))
    y_length=round(max(contour[:,0]))-round(min(contour[:,0]))
    if(y_length==0 or x_length==0):
        continue
    if(x_length/y_length>=1.5 and x_length/y_length<=3.5):
         bounding_boxes.append([round(min(contour[:,1])),round(max(contour[:,1]))
                            ,round(min(contour[:,0])),round(max(contour[:,0]))])
            
img_with_boxes=origenal_image_gray.copy()
#When provided with the correct format of the list of bounding_boxes, this section will set all pixels inside boxes in img_with_boxes
for box in bounding_boxes:
    [Xmin, Xmax, Ymin, Ymax] = box
    rr, cc = rectangle(start = (Ymin,Xmin), end = (Ymax,Xmax), shape=origenal_image_gray.shape)
    img_with_boxes[rr, cc] = 1 #set color white
show_images([origenal_image_gray,img_with_boxes],['original','boxed'])
```

```python
'''
3) Skeletonization:
    - Load "horse.jpg" from the img folder 
    - Use a binary threshold of 0.9 to remove invisible noise in the background
    - Use Skimage's "skeletonize(image)" to acquire the image's skeleton & show it
    - Use Skimage's "thin(image, max_iter)" with values of max_iter 5/10/15/20 & show it
'''
origenal_image=io.imread('./img/horse.jpg')

origenal_image_gray=gray_scale(io.imread('./img/horse.jpg'))
# print(origenal_image_gray.shape[0])
origenal_image_binary=origenal_image_gray.copy()
origenal_image_binary[origenal_image_binary<.9]=0
origenal_image_binary[origenal_image_binary>=.9]=1
skelton_image=skeletonize(origenal_image_binary)
thin_image05=thin(origenal_image_binary, 5)
thin_image10=thin(origenal_image_binary, 10)
thin_image15=thin(origenal_image_binary, 15)
thin_image20=thin(origenal_image_binary, 20)

show_images([origenal_image,origenal_image_gray,origenal_image_binary],['origenal','gray scaled','binary'])
show_images([skelton_image,thin_image05,thin_image10,thin_image15,thin_image20],['skelton_image','thin_image05','thin_image10','thin_image15','thin_image20'])
```
