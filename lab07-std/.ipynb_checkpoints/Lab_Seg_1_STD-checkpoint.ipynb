```python
def gray_scale(image):
    gray_scale_image=image
    if(len(image.shape)==3):
        if(image.shape[2]==4):
            gray_scale_image=rgb2gray(rgba2rgb(image))
        else :
            gray_scale_image=rgb2gray(image)
    else :
        gray_scale_image=gray_scale_image/255
    return gray_scale_image
```

```python
## For all requirements: The quality of the results is evaluated.


# 1- Read the image 
# [VERY IMPORTANT] and convert it to float: image = image.astype(float)

# read image golf
image=io.imread('./imgs/exp1/golf.jpeg')
gray=gray_scale(image)
image=image.astype(float)

 # 2- Retreive the 3 channels of the image(R,G,B) (will be used in the coming steps)
R_channel=image[:,:,0]
G_channel=image[:,:,1]
B_channel=image[:,:,2]
show_images([gray,image/255,R_channel,G_channel,B_channel])
# 3- Visually (not by code) detect the dominating color channel (C) for the backgroung (R, G or B)
#--> the green channel is the dominate

# 4- Draw the dominating channel (r, g or b) thresholded by some threshold (to make sure your choice is meaningful)
# G_channel=np.where(((image[:,:,1]>image[:,:,0]) & (image[:,:,1]>image[:,:,2])),255,G_channel)

# 
#the threshold we choose is multi level threshold ,
# 0 -> 0.45 will be zero
# .45-> 0.8 will be .5
# .8 -> 1 will be 1
# why?
lower_thersh=.45
uper_thersh=0.8
G_channel=G_channel/255



R_channel=R_channel/255
# R_channel=np.where(((R_channel>=lower_thersh)& (R_channel<uper_thersh) ),.5,R_channel)
R_channel=np.where(((R_channel>=lower_thersh)),1,0)
# R_channel=np.where(((R_channel<lower_thersh)),0,R_channel)
show_images([R_channel])



# G_channel=np.where(((G_channel>=lower_thersh)& (G_channel<uper_thersh) ),.5,G_channel)
G_channel=np.where(((G_channel>=lower_thersh)),1,0)
# G_channel=np.where(((G_channel<lower_thersh)),0,G_channel)
show_images([G_channel])



B_channel=R_channel/255
# B_channel=np.where(((B_channel>=lower_thersh)& (B_channel<uper_thersh) ),.5,B_channel)
B_channel=np.where(((B_channel>=lower_thersh)),1,0)
# B_channel=np.where(((B_channel<lower_thersh)),0,B_channel)
show_images([R_channel])
# 5- Get an intuitive way(There is a very simple way) to make a mask that can get the pixels where the values of the dominating channels 
# have larger values than the other channels with an appropriate threshold
gray=1
gray=np.where(((image[:,:,1]>image[:,:,0]) & (image[:,:,1]>image[:,:,2])),0,gray)
# image[:,:,1]=0
# 6- display the binarized image
show_images([gray])



```

```python
# 1- Get the required RGB color (Using a color picker)
# color: #53682E
# R= 83 G =104  B =46


# 2- Read image
image=io.imread('./imgs/exp2/cufe.png')
# 3- extract R, G and B channels (as float)

R_chennal=image[:,:,0].astype(float)
G_chennal=image[:,:,1].astype(float)
B_chennal=image[:,:,2].astype(float)
show_images([image,R_chennal,G_chennal,B_chennal])

# 4- calculate differences FOR EACH CHANNEL (between the image and the required pixel value)
dif_R_chennal=83-R_chennal
dif_G_chennal=104-G_chennal
dif_B_chennal=46-B_chennal

# 5- calculate overall distance from the given RGB color (use any appropriate distance measure, take care of the minus distances)
overall_diff=np.round((dif_R_chennal**2 + dif_G_chennal**2  + dif_B_chennal**2 )**(.5))
# io.imshow((dif_R_chennal))
# io.show()
# io.imshow(dif_G_chennal)
# io.show()
# io.imshow(dif_B_chennal)
# io.show()
# io.imshow(overall_diff)
# io.show()
# 6-  create a mask by thresholding the differences
show_images([overall_diff,dif_R_chennal,dif_G_chennal,dif_B_chennal])
mask=np.where((overall_diff>0) & (overall_diff<30),True,False)
# 7- In the input image, Replace the pixels of the mask with the following color
# R = 230 , G = 90 , B=40
image[:,:,0]=np.where(((mask==1)),230,image[:,:,0])
image[:,:,1]=np.where(((mask==1)),90,image[:,:,1])
image[:,:,2]=np.where(((mask==1)),40,image[:,:,2])
# 8- show the image or save it
show_images([image])
```
